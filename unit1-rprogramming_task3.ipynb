# 1. Calculate the following
a. Σ with and without loop.

b. Calculate U40 which is a series, Un = U(n-1) + U(n-2) , with U1= 1 and U2 = 2 ?

c. Write a program to calculate cos x, sin x(Taylor series) , Get the input(x) from the user and n=45.

d. Create a vector of the values of e cos(x) at x = 3, 3.1, 3.2, . . . , 6.

e. x <- 1:300. How many numbers in x are divisible by 2? (Use the modulo operator : %% )

```r
# a. Σ with loop
s <- 0
for (n in 5:100)
{
    s <- s + (n^4+5*n^3)
}
s
```

```r
# a. Σ without loop
n = 5:100
sum(n^4+5*n^3)
```

```r
# b. Calculate U40 which is a series, Un = U(n-1) + U(n-2) , with U1= 1 and U2 = 2
n = 40
U = 1:n
Un <- U[(n-1)]+U[(n-2)]
Un
```

```r
# c. Write a program to calculate sin x(Taylor series) , Get the input(x) from the user and n=45.
library(pracma)
x <- as.numeric(readline(prompt="Enter x: "))
n = 45
taylor(sin, x, n=4)
```

```r
# c. Write a program to calculate cos x(Taylor series) , Get the input(x) from the user and n=45.
library(pracma)
x <- as.numeric(readline(prompt="Enter x: "))
n = 45
taylor(cos, x, n=4)
```

```r
# d. Create a vector of the values of e^x cos(x) at x = 3, 3.1, 3.2, . . . , 6.
x <- seq(3, 6, by = 0.1)
x <- exp(x)*cos(x)
x
```

```r
# e. x <- 1:300. How many numbers in x are divisible by 2? (Use the modulo operator : %% )
x <- 1:300
c = 0
for (i in x)
{
    if (i%%2 == 0 )
    {
        c = c + 1
    }
}
c
```

# 2. Solve the following system of linear equations using Gaussian elimination (Ax=y)

x + 2y + 3x = 9

2x - y + 2z = −3

3x + 3y + z = 5

```r
# install matlib library
install.packages("rgl", repos="http://R-Forge.R-project.org")
install.packages("matlib")
```

```r
library(matlib)
A <- matrix(c(1, 2, 3,
              2, 1, 2,
              3, 3, 1), 3, 3, byrow=TRUE)
b <- c(9, -3, 5)
result <- gaussianElimination(A, b)
result
paste("Result is x =", result[10], ' y = ', result[11], ' z =', result[12])
```

# 4. Get the COVID-19 Dataset from the data sources. Number of observations should be more than 100.Then, report the following informations

a. Data Source detail(Ex: Link)

c. Find the missing values(rows & columns) and replace them with mean(Tidy Dataset)

d. Generate the two new variables(Var1:Mean, Var2: Median from available variable)

f. Create a plot using following instructions (using 7 layers of Grammar of Graphics)

    i. Choose x and y axis(aes) 

    ii. geom_point() - specify the parameters, size : 5, color: red, alpha: ⅕ 

    iii. Use Facet grid, cartesian coordinates & geom_smooth()

    v. Export the graph to your working directory with the title called “covid_19_ dataset.png”

```r
# a. Data Source detail(Ex: Link) 
# Dataset on Novel Corona Virus Disease 2019 in India
# Author: sudalairajkumar
# URL : https://www.kaggle.com/sudalairajkumar/covid19-in-india
library(tidyverse)
dataset <- read_csv("../input/covid19-in-india/StatewiseTestingDetails.csv")
covidDataSet <- dataset
covidDataSet
```

```r
# b. Explain the Unit & Necessity of each variable 
# This dataset have 5 variables:
# - Date: Date of observation
# - State: State in India
# - TotalSamples: Cumulative number of total samples tested till the given date
# - Negative: Cumulative number of negative samples till the given date
# - Positive: Cumulative number of positive samples till the given date

names(covidDataSet)
```

```r
# c. Find the missing values(rows & columns) and replace them with mean(Tidy Dataset) 
covidDataSet[] <- lapply(covidDataSet, function(x) { 
  x[is.na(x)] <- mean(x, na.rm = TRUE)
  x
})
covidDataSet
```

```r
# d. Generate the two new variables(Var1:Mean, Var2: Median from available variable) 
covidDataSet %>% mutate(Mean = mean(Positive),Median = median(Positive))
```

```r
# e. Rename the two existing variables 
covidDataSet %>% rename(N = Negative, P = Positive)
```

```r
# f. Create a plot using following instructions (using 7 layers of Grammar of Graphics)
```

```r
# i. Choose x and y axis(aes) 
# Choose Date as X and Positive case of covid as Y
data <- dataset %>% group_by(Date) %>% count(Date, wt = Positive)
Date <- data[['Date']]
Positive <- data[['n']]
data
```

```r
# ii. geom_point() - specify the parameters, size : 5, color: red, alpha: ⅕ 
library(ggplot2)
gp = geom_point(alpha = 1/5, color = "red", size = 5)
gp
p <- ggplot(x = data, mapping = aes(x = Date, y = Positive)) + gp
p
```

```r
# iii. Use Facet grid, cartesian coordinates & geom_smooth() 

# facet_grid
p + facet_grid(Positive)
```

```r
# coord_cartesian
p + coord_cartesian(expand = FALSE)
```

```r
#geom_smooth
p + geom_smooth()
```

```r
# iv. Assign the title to x, y and graph 
p <- p + ggtitle("Number of Positive Covid19 in India") +
  xlab("Date") + ylab("Number of Positive Cases")
p
```

```r
# v. Export the graph to your working directory with the title called “covid_19_ dataset.png”
ggsave(filename = "/covid_19_dataset.png", units = "cm", width = 25, height = 25)
```
